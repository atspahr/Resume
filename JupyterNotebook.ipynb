```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

#Using Kaggle and a linked journal I downloaded the .csv files
#https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0222112

```python
#read csv files into data frame, delimiter is semicolon
df1 = pd.read_csv('DatabaseMiniPons.csv', sep = ";")
df2 = pd.read_csv('journal.pone.0222112.s001.csv', sep = ";")
```

```python
#subscales, gender column is missing in df1
df1.head(20)
```

```python
#This dataframe has the gender
df2.head(10)
```

```python
#this steals the Sexo column from df2 and appends it to df1
df1["Sex"] = df2.Sexo.apply(lambda x: 'Male' if x== "Hombre" else "Female")
df1.head(20)
```

```python
#describe produces descriptive statistics
dsdf1 = df1.describe(include="all")
dsdf1.head(20)
```

```python
#group by Sex
sexMean = df1.groupby(['Sex']).mean()
sexMean.head()
```

```python
#group by other columns
sexMean2 = df1.groupby(['Group']).mean()
sexMean2.head()
```

```python
#query unique data in a column
df1["Group"].unique()
```

```python
#use agg function to make summary data from columns
summary = df1.groupby(['Type',"Sex"]).agg(['mean', 'std']).round(2)

#the line below will save to csv
summary.to_csv("Kaggle_Summary_Stats")

summary.head(10)
```

```python
#alternatively use for loop for summary columns

for i in pd.unique(df1.Type):
    print("----" + i + "----")
    print(df1[df1.Type == i].agg(["mean", "std"]).round(2).Right_answers)
```

```python
#Using seaborn, plot some of the data

#settin style of background
sns.set(style="ticks")

#creating custom palette
custom_palette = ["orange", "blue"]
sns.set_palette(custom_palette)

#using a catplot and designating what to plot
g = sns.catplot(x="Type", y="Age", hue="Sex", kind="swarm", data=df1)

#Formatting figure
g.fig.suptitle("Mental Health and Age", y=1)
g.set_xticklabels(['Bipolar Disorder 1', 'Bipolar Disorder 2', 'Control', 'Unipolar Depression'])
g.set(xlabel="Diagnosis")
plt.xticks(rotation=90)

plt.show()
g.savefig('figure.png', dpi=800)
```
